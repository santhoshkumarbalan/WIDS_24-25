```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x, y):
    return x**2 + y**2

# Create a grid of x and y values
x = np.linspace(-5, 5, 100)  # Range for x
y = np.linspace(-5, 5, 100)  # Range for y
X, Y = np.meshgrid(x, y)  # Create a 2D grid
Z = f(X, Y)  # Compute the function values

# Plot the level curves
plt.figure(figsize=(8, 6))
contour = plt.contour(X, Y, Z, levels=10, cmap='viridis')  # 10 levels
plt.clabel(contour, inline=True, fontsize=8)  # Label the contours
plt.title("Level Curves of f(x, y) = x² + y²")
plt.xlabel("x")
plt.ylabel("y")
plt.colorbar(contour, label="Function Value")  # Optional colorbar
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def g(x):
    return x[0]**2 + x[1]**2
```

```python
x_arange = np.arange(-10,10,0.1)
y_arange = np.arange(-10,10,0.1)
```

```python
X,Y = np.meshgrid(x_arange,y_arange)
Z = g([X,Y])
```

```python
plt.figure(figsize=(8,10))
contour = plt.contour(X,Y,Z,levels =25)
plt.clabel(contour, inline = True, fontsize = 10)
plt.show()
```
